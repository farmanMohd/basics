# isin() function

```python
mask=data["over"]>=16
```

```python
death_over=data[mask]
```

```python
death_over
```

```python
all_batsman=death_over.groupby("batsman")["batsman"].count()
x=all_batsman>200
batsman_list=all_batsman[x].index.tolist()
```

```python
batsman_list
```

```python
mask=death_over["batsman"].isin(batsman_list)
final=death_over[mask]
```

```python
sumVal=final.groupby("batsman")["batsman_runs"].sum()
sumVal
```

```python
balls=final.groupby("batsman")["batsman_runs"].count()
```

```python
sr=(sumVal/balls)*100
```

```python
sr=sr.sort_values(ascending=False)
sr
```

```python
sr.index[0]
```
